```python
from qiskit import *
from qiskit.visualization import plot_histogram, array_to_latex, plot_bloch_multivector,plot_bloch_vector
%matplotlib inline
```

```python
qr = QuantumRegister(1)
cr = ClassicalRegister(1)
circuit = QuantumCircuit(qr,cr)
circuit.h(0)
circuit.measure(qr,cr)
```
